# Training and testing data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(iris_data['data'],iris_data['target'],random_state=0)

```

```python
#data column in x_train 75%
print(x_train)
#target column in y_train 75%
print(y_train)
```

# Look at our data

```python
#create a dataframe from data x_train 
#labels the column using the string in iris_data.features_name

iris_frame=pd.DataFrame(x_train,columns=iris_data.feature_names)
print(iris_frame)
```

```python
#create a scatter matrix
from pandas.plotting import scatter_matrix
scatter_matrix(iris_frame, alpha=0.2, figsize=(10, 10))
plt.show()
```

# Building first Model -K - Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=1)

knn.fit(x_train,y_train)
```

# Making  predictions

```python
new_iris=([[1.2,3,3.0,2]])

prediction=knn.predict(new_iris)

print("new Iris prediction:{}\n".format(prediction))
print("Predicted target name: {}".format(
iris_data['target_names'][prediction]))
```

# Evaluting Model

```python
y_pred=knn.predict(x_test)

print("Test set predictions:\n {}".format(y_pred))

print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```
